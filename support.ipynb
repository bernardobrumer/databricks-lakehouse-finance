```python
from pyspark.sql.functions import current_timestamp, col

def add_ingestion_timestamp(input_df):
    input_df = input_df.withColumn('ingestion_timestamp', current_timestamp())
    return input_df
```

```python
def add_file_metadata(input_df):
    input_df = input_df.withColumn('file_name', col('_metadata.file_name')) \
        .withColumn('file_path', col('_metadata.file_path'))
    return input_df
```
