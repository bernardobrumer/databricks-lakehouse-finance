```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType, TimestampType, DateType
```

```python
schema = StructType(fields=
                    [StructField('data_pagamento', DateType(), True),
                    StructField('data_vencimento', DateType(), True),
                    StructField('status', StringType(), True),
                    StructField('valor_brl', DoubleType(), True)]
                    )
```

```python
display(dbutils.fs.ls('/Volumes/personalfinance/landing/vol_landing/boletos/'))
```

```python
df = spark.read \
    .format('json') \
    .schema(schema) \
    .option('multiLine', 'true') \
    .load('/Volumes/personalfinance/landing/vol_landing/boletos/*.json')
```

```python
df = add_file_metadata(df)
df = add_ingestion_timestamp(df)
```
